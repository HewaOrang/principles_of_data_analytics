# matplotlib

```python
# Plots.
import matplotlib.pyplot as plt

# Numerical arrays.
import numpy as np
```

# Plotting Functions

```python
# X values.
x = np.linspace(0.0, 10.0, 101)

# Show.
x
```

```python
# Y values.
y = x**2

# Show.
y
```

```python
# Plot x vs y.
plt.plot(x, y, "k-", label='$y = x^2$')

# Add a legend.
plt.legend()

# Add labels.
plt.xlabel('x')
plt.ylabel('$x^2$')

# Add a title.
plt.title('A Lovely Plot');
```

# Cpmparing Plots

```python
# Domain of x values
x = np.linspace(0.0, 100, 1001)

# x squared.
y1 = x**2 

# 2 to the power of x.
y2 = 2**x
```

```python
# Plot x vs y1
plt.plot(x, y1, label= '$x^2$')

# Plot x vs y2.
plt.plot(x, y2, label='$2^x$')

# Add a legend.
plt.legend()

# Add labels
plt.xlabel('x')
plt.ylabel('y')

# Add a title
plt.title('Two loverly plots')
```
