## Fit a Line

```python
# Range of x values.
x = np.linspace(-10.0, 10.0, 21)

# Range of y values.
y= 4.0 * x + 2.0 + np.random.normal(0.0, 3.0, x.shape)
```

```python
# plot x and y.
plt.plot(x, y, 'o', label='data')
```

```python
# Fit a line.
m, c = np.polyfit(x, y, 1)

# Show.
m, c
```

```python
# plot x and y.
plt.plot(x, y, 'o', label='data')

# plot the line.
plt.plot(x, m*x + c, label='fit')
```

```python
# Fit a line.
coeffs = np.polyfit(x, y, 1)
# Show.
coeffs
```

```python
# Evaluate the polynomial.
f = np.polyval(coeffs, x)

# show.
f
```

## Evaluate
https://en.wikipedia.org/wiki/Coefficient_of_determination

```python
# Use scipy to fit a line.
fit = sp.stats.linregress(x, y)
# fit slope
fit.slope
# fit intercept
fit.intercept
# R value
fit.rvalue
# R^2 value
fit.rvalue**2
```
